```python
##------- Prepare data files for analysis -------------##
# Output csv file of GeomeTRe and json files of the RepeatsDB database used to get
# classification and repeat unit number information from RepeatsDB. .json file can be retrieved via API (https://repeatsdb.org/api)
# test files in test_data directory

import pandas as pd
import json

# Load GeomeTRe CSV file
geometre_df = pd.read_csv("test_data/test_output_geometre.csv")

# Load RepeatsDB annotations JSON
with open("test_data/test_repeatsdb4-annotations.json", "r") as f:
    repeats_data = json.load(f)

# Create a dictionary from JSON based on structure and chain
repeat_lookup = {}
for entry in repeats_data:
    content = entry["content"]
    structure = content["chain"]["structure"].lower()
    chain_id = content["chain"]["id"]
    region_class = content["region_classes"][0] if content["region_classes"] else None
    repeat_units = sum(1 for l in content["loci"] if l["type"] == "unit")
    repeat_lookup[(structure, chain_id)] = {
        "classification": region_class,
        "repeat_unit_number": repeat_units
    }

# Annotate DataFrame
geometre_df["classification"] = geometre_df.apply(
    lambda row: repeat_lookup.get((row["pdb_id"].lower(), row["chain"]), {}).get("classification", None),
    axis=1
)
geometre_df["repeat_unit_number"] = geometre_df.apply(
    lambda row: repeat_lookup.get((row["pdb_id"].lower(), row["chain"]), {}).get("repeat_unit_number", None),
    axis=1
)

# Preview
print(geometre_df.head())

# Save the annotated DataFrame to CSV file
output_path = "output_path/geometre_annotated_output.csv"
geometre_df.to_csv(output_path, index=False)
```

```python
# Figure S9-S11 correlation between mean and standard deviation in curvature, twist, and pitch
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Load the STRPs data
strps_input = "../data/geometre_annotated_output.csv"
strps_df = pd.read_csv(strps_input)

# Strip any whitespace from column names to ensure consistency
strps_df.columns = strps_df.columns.str.strip()

# Convert relevant columns to numeric, coercing any non-numeric values to NaN
strps_df['pitch_mean'] = pd.to_numeric(strps_df['pitch_mean'], errors='coerce')
strps_df['pitch_std'] = pd.to_numeric(strps_df['pitch_std'], errors='coerce')
strps_df['classification'] = strps_df['classification'].astype(str)  # Convert 'CTFC' to string for consistent filtering

# Drop rows with NaN values in 'curv_mean' or 'curv_std'
strps_df.dropna(subset=['pitch_mean', 'pitch_std'], inplace=True)

# Filter out rows where 'classification' is NaN, '2.1', or '2.2'
strps_df = strps_df[~strps_df['classification'].isin(['nan', '2.1', '2.2','4.7'])]
strps_df = strps_df[~strps_df['classification'].str.startswith('5')]

# Extract the first two components of the classification value (e.g., "3.3" from "3.3.1.4")
strps_df['classification_group'] = strps_df['classification'].apply(lambda x: '.'.join(x.split('.')[:2]))
# Filter rows where 'classification_group' equals '3.1'
filtered_df = strps_df[strps_df['classification_group'] == '4.8']
print(filtered_df)

# Calculate global min and max for normalization
min_curv_mean = strps_df['pitch_mean'].min()
max_curv_mean = strps_df['pitch_mean'].max()
min_curv_std = strps_df['pitch_std'].min()
max_curv_std = strps_df['pitch_std'].max()

# Normalize interested columns
strps_df['pitch_mean'] = (strps_df['pitch_mean'] - min_curv_mean) / (max_curv_mean - min_curv_mean)
strps_df['pitch_std'] = (strps_df['pitch_std'] - min_curv_std) / (max_curv_std - min_curv_std)

# Group by the classification
groups = strps_df.groupby('classification_group')

# Determine the number of subplots needed
num_groups = len(groups)
cols = 4  # Number of columns in the grid
rows = (num_groups + cols - 1) // cols  # Calculate rows based on the number of groups and columns

# Create a figure with subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 4))
axes = axes.flatten()  # Flatten to easily index each subplot

# Plot each group in a separate subplot
for i, (name, group) in enumerate(groups):
    # Fig S11 color='#8c564b'
    # Fig S10 color='#17becf'
    # Fig S9 color ='#e377c2'
    axes[i].scatter(group['pitch_mean'], group['pitch_std'], alpha=0.7, color='#8c564b')
    
# Calculate and plot linear regression line
    slope, intercept, r_value, p_value, std_err = stats.linregress(group['pitch_mean'], group['pitch_std'])
    x_vals = np.array(axes[i].get_xlim())
    y_vals = intercept + slope * x_vals
    #axes[i].plot(x_vals, y_vals, color='black', linewidth=1.5, label=f'ρ = {r_value:.2f}')
    axes[i].plot(x_vals, y_vals, color='black', linewidth=1.5, label=f' p ={p_value:.3g}')
    
    # Add correlation coefficient (Pearson's r) to the plot
    axes[i].text(0.05, 0.95, f'r ={r_value:.2f}', transform=axes[i].transAxes, 
                fontsize=16, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.7))
    
    axes[i].set_title(f'{name}', fontsize=24)
    axes[i].set_xlabel('pitch mean', fontsize=14)
    axes[i].set_ylabel('pitch std', fontsize=14)
    axes[i].grid(False)
    axes[i].set_xlim(0, 1.0) 
    axes[i].set_ylim(0, 1.0)  

# Remove any empty subplots if the number of groups does not perfectly fill the grid
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
# Show the figure
plt.show()
```

```python
# Main figure D t-SNE plot
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = '../data/geometre_annotated_output.csv'
data = pd.read_csv(file_path, sep=',')

# Step 2: Clean up column names
data.columns = data.columns.str.strip()

# Step 3: Create unique ID and extract fold-level category (e.g., "3.1.1" from "3.1.1.4")
data['id'] = data['pdb_id'] + '_' + data['chain']
data['CTFC_category'] = data['classification'].astype(str).str.split('.').str[:3].str.join('.')

# Step 4: Define and filter by fold-level categories
custom_colors = {
    '3.1.1': '#0000FF',  # Blue
    '3.1.2': '#008000',  # Green
    '3.2.1': '#FF0000',  # Red
    '3.2.2': '#800080',  # Purple
    '3.3.1': '#FA8072',  # Salmon
    '3.3.2': '#FFA500',  # Orange
    '3.3.3': '#FFD700',  # Yellow
    '3.4.1': '#00FF00',  # Bright Green
    '3.4.2': '#5DADE2',  # Sky Blue
}
data = data[data['CTFC_category'].isin(custom_colors.keys())]

# Step 5: Select numerical features and normalize
numerical_columns = [
    "curvature_mean", "curvature_std", "twist_mean", "twist_std",
    "pitch_mean", "pitch_std"
]
numerical_data = data[numerical_columns]
scaler = StandardScaler()
normalized_data = scaler.fit_transform(numerical_data)

# Step 6: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=50, max_iter=3000, random_state=40)
tsne_result = tsne.fit_transform(normalized_data)

# Step 7: Prepare colors and output DataFrame
point_colors = data['CTFC_category'].map(custom_colors)
tsne_df = pd.DataFrame(tsne_result, columns=['Component 1', 'Component 2'])
tsne_df['id'] = data['id'].values
tsne_df['CTFC_category'] = data['CTFC_category'].values

# Step 8: Save t-SNE data
output_csv_path = '/output_path/values.csv'
tsne_df.to_csv(output_csv_path, index=False)

# Step 9: Plot
plt.figure(figsize=(8, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=point_colors, s=10, alpha=0.5)

# Step 10: Add legend
handles = [
    plt.Line2D([0], [0], marker='o', color=color, linestyle='', label=label)
    for label, color in custom_colors.items() if label in data['CTFC_category'].unique()
]
plt.legend(handles=handles, title="Topology", bbox_to_anchor=(1.05, 1), loc='upper left')

# Final touches
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.tight_layout()
plt.show()
```

```python
# Figure S4-S5 Twist and pitch handedness in Class 3 and Class 4 with repeat unit number >=6
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
input_file = "../data/geometre_annotated_output.csv"  
strps_df = pd.read_csv(input_file)

# Clean column names
strps_df.columns = strps_df.columns.str.strip()

# Convert columns to numeric
strps_df['twist_hand_mean'] = pd.to_numeric(strps_df['twist_hand_mean'], errors='coerce')
strps_df['classification'] = strps_df['classification'].astype(str)
strps_df['twist_mean'] = pd.to_numeric(strps_df['twist_mean'], errors='coerce')
strps_df['twist_std'] = pd.to_numeric(strps_df['twist_std'], errors='coerce')

# Drop NaN values
strps_df.dropna(subset=['twist_hand_mean', 'twist_mean', 'twist_std'], inplace=True)

# Filter for classes starting with "3" or "4"
strps_df = strps_df[strps_df['classification'].str.startswith(('3', '4'))]
strps_df['classification_group'] = strps_df['classification'].apply(lambda x: '.'.join(x.split('.')[:2]))

# Group by topology
groups = strps_df.groupby('classification_group')

# Apply filtering to each group based on conditions
filtered_groups = []
for name, group in groups:
    # Filter the group based on twist/pitch_mean > 0.1 and twist/pitch_std < 0.1
    filtered_group = group[(group['pitch_mean'] > 0.1) & (group['pitch_std'] < 0.1) & (group['repeat_unit_number'] > 5)]
    #print(filtered_group)
    if not filtered_group.empty:
        filtered_groups.append(filtered_group)

# Combine filtered groups
filtered_strps_df = pd.concat(filtered_groups)

# Group the filtered DataFrame
filtered_groups = filtered_strps_df.groupby('classification_group')

# Set number of plots
num_groups = len(filtered_groups)
cols = 4
rows = (num_groups + cols - 1) // cols  

# Create figure
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 4))
axes = axes.flatten()

# Define fixed bins with edge alignment
shared_bins = np.linspace(-1, 1, 21)  # 20 bins of width 0.1

# Plot each histogram
for i, (name, group) in enumerate(filtered_groups):
    total_regions = len(group)
    
    # Force histogram to use full range
    counts, bin_edges = np.histogram(
        group['twist_hand_mean'],
        bins=shared_bins,
        range=(-1, 1)  # Explicit range enforcement
    )
    
    # Convert to percentage with zero-division handling
    percentages = (counts / total_regions) * 100 if total_regions != 0 else np.zeros_like(counts)
    
    # Plot with edge alignment
    axes[i].bar(
        bin_edges[:-1], 
        percentages, 
        width=np.diff(bin_edges),
        color='#2E8B57', 
        alpha=0.7, 
        edgecolor='black',
        align='edge' 
    )

    # Plot histogram using calculated percentages
    # Fig S5 color ='#BC8F8F'
    # Fig S4 color ='#2E8B57'
    
    # Consistent formatting
    axes[i].set_title(f'{name}', fontsize=24)
    axes[i].set_xlabel('twist handedness mean', fontsize=10)
    axes[i].set_ylabel('Percentage (%)', fontsize=10)
    axes[i].set_ylim(0, 100)
    axes[i].set_xlim(-1, 1)  # Match bin range
    axes[i].grid(False)

# Remove extra empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()

# Show the figure
plt.show()
```
